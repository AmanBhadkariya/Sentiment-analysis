```python
pos_words = open('positiveWords.txt','r').read()
pos_words = pos_words.split()
neg_words = open('negativeWords.txt','r').read()
neg_words = neg_words.split()

def pw_ratio(text):
    if len(text) == 0:
        return 0
    pos_count = 0
    t = text.lower().split()
    for word in t:
        if word in pos_words:
            pos_count += 1
    return pos_count/len(text)

def nw_ratio(text):
    if len(text) == 0:
        return 0 
    neg_count = 0
    t = text.lower().split()
    for word in t:
        if word in neg_words:
            neg_count += 1
    return neg_count/len(text)

def pw(text):
    if len(text) == 0:
        return 0
    pos_count = 0
    t = text.lower().split()
    for word in t:
        if word in pos_words:
            pos_count += 1
    return pos_count

def nw(text):
    neg_count = 0
    t = text.lower().split()
    for word in t: 
        if word in neg_words:
            neg_count += 1
    return neg_count

def dif_pn(text):
    neg_count = 0
    t = text.lower().split()
    for word in t:
        if word in neg_words:
            neg_count += 1
    pos_count = 0
    t = text.lower().split()
    for word in t:
        if word in pos_words:
            pos_count += 1
    return pos_count - neg_count
```

```python
d['pos_ratio'] = d['text'].apply(pw_ratio)
d['neg_ratio'] = d['text'].apply(nw_ratio)
d['pos_cnt'] = d['text'].apply(pw)
d['neg_cnt'] = d['text'].apply(nw)
d['dif_pos_neg'] = d['text'].apply(dif_pn)
```
